### Problem 2

#### Problem Statement:
Plot a line graph on income vs year for the state of Arizona

```python
import numpy as np

t=np.arange(0.1,2.0,0.1)
s=1+np.sin(2*np.pi*t)    
plt.xlabel('Time')       #Give a X axis label name
plt.ylabel('Voltage')    #Give a Y axis label name
plt.plot(t,s)            #Used to plot the t,s co ordinates
plt.title('Flying Car')  #Used to display a title above the graph
plt.grid(True)           #Used to display grids in the graph
plt.savefig('Figures/TimeVoltage')  #Used to save the graph as a png image

plt.show()
```

### Problem 3

#### Problem Statement:
Line Graph for Income of Arizona

```python
import pandas as pd

def incomeplot(inc_filename):
    df=readCSV(inc_filename)
    headings=df.columns     #it gives all the first column values
    li=df.values[2]         #it gives all the values present in 2nd row(Arizona)
    
    
    years=[ ]               #Empty List
    income=[ ]              #Empty List
    for i in range(2,len(headings)):
        years.append(headings[i])
    for i in range(2,len(li)):
        income.append(li[i])
    #x=np.arange(years)
    #y=np.arange(income)
    plt.xlabel("Years")
    plt.ylabel("Income")
    plt.title("Income in Arizona")
    plt.plot(years,income)
    plt.grid(True)
    plt.show()
    return years
    return income

def readCSV(inc_filename):
    df=pd.read_csv(inc_filename)
    return df


incomeplot('DataFiles/Income.csv')
```

```python
width=1/2
incomeplot('DataFiles/Income.csv')
plt.bar(years,income,width,color='black')
plt.title("Bar Cart for Income of Arizona")
plt.show()
```

### Problem 4

#### Problem Statement:
Generate Scatter Plot for 50 different numbers

```python
n=50
x=np.random.rand(n)      #Generate 50 random numbers
y=np.random.rand(n)      #Generate 50 random numbers
colorlist=np.random.rand(n)        #Generate 50 random colors
arealist=np.pi*(15*np.random.rand(50))**2          #Generate 50 random areas of circles
plt.scatter(x,y,s=arealist,c=colorlist,alpha=10)

plt.savefig('Figures/Scatter-Plot-50-random-circles')
```
