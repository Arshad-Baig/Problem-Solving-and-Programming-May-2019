### Problem 2

#### Problem Statement:
Define a function to add new row data to a CSV file.

```python
def addRowCSV(filename,rowdata):
    df=readCSV(filename)
    df.loc[len(df)]=rowdata            #location of where to add the new row data
    df.to_csv(filename, index=False)
    return df 
rowdata=['010101','statename','1','2','3','4','5','6','7','8','9'] #Creating new row data
addRowCSV('DataFiles/Income.csv',rowdata)
```

### Problem 3
#### Problem Statement:
Define a function to update/modify a row in a CSV File.


```python
def updateRowCSV(filename,rowindex,rowdata):
    df=readCSV(filename)
    df.loc[rowindex]=rowdata
    df.to_csv(filename,index=False)
    return df

rowdata = ['04000US08','Florida',123,4325,4565,9084,83123,423,987,345,765]
updateRowCSV('DataFiles/Income.csv', 6, rowdata)
###df=df.drop('Unnamed:0','axis=1')
#rowdata=['010101','statename','1','2','3','4','5','6','7','8','9']
#updateRowCSV('DataFiles/Income.csv',6,rowdata)
```

### Problem 4

#### Problem Statement:
Define a program to delete a row from a CSV File.

```python
def deleteRowCSV(filename,rowindex):
    df=readCSV(filename)
    df=df.drop(rowindex)
    df.to_csv(filename,index=False)
    return df
```

```python
filename='DataFiles/Income.csv'
rowindex=6
deleteRowCSV(filename,rowindex)
```
